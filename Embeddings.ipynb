```python
df = pd.DataFrame(documents)
df.columns = ['Titulo', 'Conteudo']
df
```

```python
model = 'models/embedding-001'
```

```python
def embad_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type='RETRIEVAL_DOCUMENT')['embedding']
```

```python
df['Embeddings'] = df.apply(lambda row: embad_fn(row['Titulo'], row['Conteudo']), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type='RETRIEVAL_QUERY')['embedding']

  produtos_escalares = np.dot(np.stack(df['Embeddings'].ravel()), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)

  return df.iloc[indice]['Conteudo']
```

```python
consulta = 'Como faço para trocar a marcha do carro do google?'

trecho = gerar_e_buscar_consulta(consulta, df, model)

print(trecho)
```

```python
generation_config = {
    'temperature': 0,
    'candidate_count': 1,
}
```

```python
prompt = f'Reescreva este texto de uma forma mais descontraída, sem adicionar informações que não fazem parte do texto {trecho}'

model_2 = genai.GenerativeModel('gemini-1.0-pro',
                                generation_config=generation_config)

response = model_2.generate_content(prompt)

print(response.text)
```
